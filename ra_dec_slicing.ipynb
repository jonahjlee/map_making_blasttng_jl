```python
import numpy as np
import matplotlib.pyplot as plt

ra_dec = np.load('ra_dec.npy', allow_pickle=True).item()
```

```python
plt.plot(ra_dec['time'], ra_dec['ra'], color='blue', label='RA')
plt.plot(ra_dec['time'], ra_dec['dec'], color='red', label='Dec')
plt.xlabel('Unix Time (s)')
plt.ylabel('Angle (deg)')
plt.legend(loc='best')
plt.show()

plt.plot(ra_dec['ra'], color='blue', label='RA')
plt.plot(ra_dec['dec'], color='red', label='Dec')
plt.xlabel('Index')
plt.ylabel('Angle (deg)')
plt.legend(loc='best')
plt.show()

plt.plot(ra_dec['time'][20000:24000], ra_dec['ra'][20000:24000], color='blue')
plt.plot(ra_dec['time'][20000:24000], ra_dec['dec'][20000:24000], color='red')
plt.xlabel('Unix Time (s)')
plt.ylabel('Angle (deg)')
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(ra_dec['ra'], ra_dec['dec'])
plt.title('RA vs. Dec., full array')
plt.xlabel('RA (deg)')
plt.ylabel('Dec. (deg)')
plt.show()
```

```python
start_index = 37200
end_index = start_index+20
plt.plot(ra_dec['ra'][start_index:end_index], ra_dec['dec'][start_index:end_index])
plt.title('RA vs. Dec., Roach 1 slice (from mmi_config.py)')
plt.xlabel('RA (deg)')
plt.ylabel('Dec. (deg)')
plt.show()

print(ra_dec['time'][start_index])
print(ra_dec['time'][end_index])
```
